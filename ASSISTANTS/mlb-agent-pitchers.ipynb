```python
import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
SERPER_API_KEY = os.getenv("SERPER_API_KEY")

# os.environ["OPENAI_MODEL_NAME"]="gpt-4o"
os.environ["OPENAI_MODEL_NAME"]="gpt-4-turbo"
# os.environ["OPENAI_MODEL_NAME"]="gpt-3.5-turbo"

search_tool = SerperDevTool()
```

```python
# List of MLB matchups and starting pitchers
upcoming_games = [
    ("Houston Astros", "Los Angeles Angels", "9:38 PM ET", "Valdez, F", "Canning"),
    ("Arizona Diamondbacks", "San Diego Padres", "9:40 PM ET", "Pfaadt", "King"),
    ("Toronto Blue Jays", "Oakland Athletics", "9:40 PM ET", "Bassitt", "Harris, H")
]

# Iterate over each game and assign to variables
for game in upcoming_games:
    team1, team2, time, pitcher1, pitcher2 = game
    print(f"Team 1: {team1}, Team 2: {team2}, Time: {time}, Pitcher 1: {pitcher1}, Pitcher 2: {pitcher2}")
```

```python
print(pitcher1)
print(pitcher2)
```

```python
agent1 = Agent(
    role="Pitcher Stats Analyst",
    goal="Analyze pitcher's performance statistics and evaluate strengths and weaknesses based on pitcher data",
    backstory="As a seasoned analyst with a deep understanding of MLB, you specialize in breaking down games and understanding pitcher tendencies and histories.",
#     max_iter=10,
    allow_delegation=True,
    tools=[search_tool]
)
```

```python
task1 = Task(
    description=f"Research detailed statistics and historical performance of {pitcher1}",
    expected_output=f"Comprehensive statistical report on {pitcher1}.",
    agent=agent1,
    output_file=f'./logs/{pitcher1}_stats.txt'
)

task2 = Task(
    description=f"Research detailed statistics and historical performance of {pitcher2}",
    expected_output=f"Comprehensive statistical report on {pitcher2}.",
    agent=agent1,
    output_file=f'./logs/{pitcher2}_stats.txt'
)    

# Create a crew with the agent and task
crew = Crew(
    agents=[agent1],
    tasks=[task1, task2],
    verbose=False
)

# Execute the task
result = crew.kickoff()
print(result)
```
