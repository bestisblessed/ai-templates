# My Agents

```python
# General Fighter Stat Analysis Agent
agent1 = Agent(
    role="Fighter Stats Analyst",
    goal="""Analyze fighter's performance statistics and evaluate strengths and
    weaknesses based on fighter data""",
    backstory=
        """As a seasoned analyst with a deep understanding of MMA/UFC, you specialize
        in breaking down fights and understanding fighter tendencies and histories.""",
    verbose=True,
    allow_delegation=True,
    tools=[search_tool],
)

# Recent Performance and Momentum Analysis Agent
agent2 = Agent(
    role="Historical Fight Analyst",
    goal="""Analyze previous fights to learn performance against similar opponents 
    as well to assess current momentum of the fighter""",
    backstory=
        """As a seasoned analyst with a deep understanding of MMA/UFC, you specialize
        in breaking down fights.
        Your expertise lies in dissecting historical fight outcomes and statistics,
        providing insights into how a fighter's recent experiences might influence their
        next performance.""",
    verbose=True,
    allow_delegation=True,
    tools=[search_tool],
)

# Public Sentiment Analysis Agent
agent3 = Agent(
    role="Public Sentiment Analyst",
    goal="""Analyze public sentiment from various online sources to gauge who the public 
    believes will win upcoming UFC fights. This helps in understanding betting trends and 
    public biases, which can inform more strategic handicapping decisions.""",
    backstory=
        """As a data-driven sentiment analyst with a focus on combat sports, you specialize
        in extracting and interpreting the mood and opinions from social media, forums, and 
        discussion boards. Your analysis provides a comprehensive view of public opinion, 
        which is crucial for predicting fight outcomes and understanding public perception 
        dynamics in the UFC.""",
    verbose=True,
    allow_delegation=True,
    tools=[search_tool],
)

# Predictor Agent
agent_final = Agent(
    role="Prediction Agent",
    goal="Predict outcomes of upcoming fights based on statistical analysis",
    backstory=
        """As a expert analyst and oddsmaker with a deep understanding of MMA/UFC,
        you specialize in breaking down and handicapping fights.
        Equipped with advanced predictive modeling and professional handicapping knowledge,
        you use and evaluate data to predict hypothetical fight outcomes with high accuracy.""",
    verbose=True,
    allow_delegation=True,
)
```

```python
# Run Tasks

# fighter1 = input("Fighter 1: ")
# fighter2 = input("Fighter 2: ")

task1 = Task(
    description=f"Research detailed statistics and historical performance of {fighter1}",
    expected_output=f"Comprehensive statistical report on {fighter1}.",
    agent=agent1,
    max_iter=10,
    output_file=f'./fight-predictions/task1_output_{fighter1}.txt'
)
task2 = Task(
    description=f"Research detailed statistics and historical performance of {fighter2}",
    expected_output=f"Comprehensive statistical report on {fighter2}.",
    agent=agent1,
    max_iter=10, 
    output_file=f'./fight-predictions/task2_output_{fighter2}.txt'
)
task3 = Task(
    description=f"Analyze historical fight data and momentum of {fighter1}",
    expected_output=f"Detailed analysis of fight history and current momentum of {fighter1}.",
    agent=agent2,
    max_iter=10,
    output_file=f'./fight-predictions/task3_output_{fighter1}.txt'
)
task4 = Task(
    description=f"Analyze historical fight data and momentum of {fighter2}",
    expected_output=f"Detailed analysis of fight history and current momentum of {fighter2}.",
    agent=agent2,
    max_iter=10,
    output_file=f'./fight-predictions/task4_output_{fighter2}.txt'
)
task5 = Task(
    description=f"Research {fighter1} versus {fighter2}",
    expected_output=f"Comprehensive analysis on the general sentiment and public biases.",
    agent=agent3,
    max_iter=10,
    output_file=f'./fight-predictions/task5_output_{fighter1}_{fighter2}.txt'
)

final_task = Task(
    description=f"""Using the insights provided, predict the outcome of a hypothetical
    fight between {fighter1} and {fighter2}.""",
    expected_output=
    """Your prediction should be informative and insightful based on findings from your 
    research. You must include the round and method of victory in your prediction.""",
    agent=agent_final,
    output_file=f'./fight-predictions/final_task_output_{fighter1}_{fighter2}.txt'
)
```

```python
# Assemble crew
crew = Crew(
    agents=[agent1, agent2, agent3, agent_final],
    tasks=[task1, task2, task3, task4, task5, final_task],
    verbose=False
)

# Execute tasks
result = crew.kickoff() 
print(result)
```

# Combine Files

```python
# Path to the directory containing the unzipped files
base_dir = './fight-predictions/ufc-louisville-2/'
output_dir = './fight-predictions/'

# Define the correct file names and group them by fight in the correct order
fight_files = {
    "Jared 'The Killa Gorilla' Cannonier_Nassourdine Imavov": [
        "task1_output_Jared 'The Killa Gorilla' Cannonier.txt",
        "task2_output_Nassourdine Imavov.txt",
        "task3_output_Jared 'The Killa Gorilla' Cannonier.txt",
        "task4_output_Nassourdine Imavov.txt",
        "task5_output_Jared 'The Killa Gorilla' Cannonier_Nassourdine Imavov.txt",
        "final_task_output_Jared 'The Killa Gorilla' Cannonier_Nassourdine Imavov.txt"
    ],
    "Dominick 'The Devastator' Reyes_Dustin 'The Hanyak' Jacoby": [
        "task1_output_Dominick 'The Devastator' Reyes.txt",
        "task2_output_Dustin 'The Hanyak' Jacoby.txt",
        "task3_output_Dominick 'The Devastator' Reyes.txt",
        "task4_output_Dustin 'The Hanyak' Jacoby.txt",
        "task5_output_Dominick 'The Devastator' Reyes_Dustin 'The Hanyak' Jacoby.txt",
        "final_task_output_Dominick 'The Devastator' Reyes_Dustin 'The Hanyak' Jacoby.txt"
    ],
    "Raul 'El Nino Problema' Rosas Jr._Ricky 'Pretty' Turcios": [
        "task1_output_Raul 'El Nino Problema' Rosas Jr..txt",
        "task2_output_Ricky 'Pretty' Turcios.txt",
        "task3_output_Raul 'El Nino Problema' Rosas Jr..txt",
        "task4_output_Ricky 'Pretty' Turcios.txt",
        "task5_output_Raul 'El Nino Problema' Rosas Jr._Ricky 'Pretty' Turcios.txt",
        "final_task_output_Raul 'El Nino Problema' Rosas Jr._Ricky 'Pretty' Turcios.txt"
    ],
    "Miguel 'Caramel Thunder' Baeza_Punahele 'Story Time' Soriano": [
        "task1_output_Miguel 'Caramel Thunder' Baeza.txt",
        "task2_output_Punahele 'Story Time' Soriano.txt",
        "task3_output_Miguel 'Caramel Thunder' Baeza.txt",
        "task4_output_Punahele 'Story Time' Soriano.txt",
        "task5_output_Miguel 'Caramel Thunder' Baeza_Punahele 'Story Time' Soriano.txt",
        "final_task_output_Miguel 'Caramel Thunder' Baeza_Punahele 'Story Time' Soriano.txt"
    ],
    "Thiago Moises_Ludovit 'Mr. Highlight' Klein": [
        "task1_output_Thiago Moises.txt",
        "task2_output_Ludovit 'Mr. Highlight' Klein.txt",
        "task3_output_Thiago Moises.txt",
        "task4_output_Ludovit 'Mr. Highlight' Klein.txt",
        "task5_output_Thiago Moises_Ludovit 'Mr. Highlight' Klein.txt",
        "final_task_output_Thiago Moises_Ludovit 'Mr. Highlight' Klein.txt"
    ],
    "Charlie 'Chuck Buffalo' Radtke_Carlos 'Carlao' Prates": [
        "task1_output_Charlie 'Chuck Buffalo' Radtke.txt",
        "task2_output_Carlos 'Carlao' Prates.txt",
        "task3_output_Charlie 'Chuck Buffalo' Radtke.txt",
        "task4_output_Carlos 'Carlao' Prates.txt",
        "task5_output_Charlie 'Chuck Buffalo' Radtke_Carlos 'Carlao' Prates.txt",
        "final_task_output_Charlie 'Chuck Buffalo' Radtke_Carlos 'Carlao' Prates.txt"
    ]
}

combined_output_files = []

# Combine the contents of the files for each fight
for fight, files in fight_files.items():
    combined_content = ""
    for file in files:
        file_path = os.path.join(base_dir, file)
        if os.path.exists(file_path):
            with open(file_path, 'r', encoding='utf-8') as f:
                combined_content += f.read() + "\n\n"
    if combined_content:
        output_file_name = fight.replace(" ", "_").replace("'", "")
        output_file_path = os.path.join(output_dir, f"{output_file_name}.txt")
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(combined_content)
        combined_output_files.append((fight, output_file_path))

# Combine all output files into a single file with separators
final_combined_path = os.path.join(output_dir, 'ufc-louisville.txt')
with open(final_combined_path, 'w', encoding='utf-8') as final_combined_file:
    for fight, file_path in combined_output_files:
        fight_title = fight.replace("_", " vs ").replace("_", " ").replace("Jr ", "Jr. ").replace("El Nino Problema", "El Niño Problema")
        final_combined_file.write(f"{fight_title}\n{'-'*80}\n")
        with open(file_path, 'r', encoding='utf-8') as infile:
            final_combined_file.write(infile.read() + "\n\n")
#         final_combined_file.write(f"{'-'*80}\n")

print(f"All files have been combined into {final_combined_path}.")
```
