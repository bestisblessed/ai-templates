```python
import requests

def search_pairs_on_chain(chain_id, query=""):
    url = f"https://api.dexscreener.com/latest/dex/search?q={query}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return [pair for pair in data['pairs'] if pair['chainId'] == chain_id]
    else:
        return None

def filter_pairs_by_volume(pairs, volume_threshold=10_000_000):
    return [pair for pair in pairs if pair.get('volume', {}).get('h6', 0) > volume_threshold]

# Example usage
chain_id = "solana"
query = " "  # Use an empty query to search broadly
pairs_data = search_pairs_on_chain(chain_id, query)

if pairs_data:
    high_volume_pairs = filter_pairs_by_volume(pairs_data)
    for pair in high_volume_pairs:
        print(pair)
else:
    print("Error: Could not fetch data or no pairs found")

```

```python
import requests
import csv

def get_solana_tokens():
    url = "https://raw.githubusercontent.com/solana-labs/token-list/main/src/tokens/solana.tokenlist.json"
    response = requests.get(url)
    if response.status_code == 200:
        token_data = response.json()
        return token_data['tokens']
    else:
        return None

# Example usage
solana_tokens = get_solana_tokens()

if solana_tokens:
    keys = solana_tokens[0].keys()  # Get the keys from the first token
    with open("solana_tokens.csv", 'w', newline='') as output_file:
        dict_writer = csv.DictWriter(output_file, fieldnames=keys)
        dict_writer.writeheader()
        dict_writer.writerows(solana_tokens)

    for token in solana_tokens:
        print(f"Token: {token['name']} ({token['symbol']}), Address: {token['address']}")
else:
    print("Error: Could not fetch data or no tokens found")
```

```python
# Scrape dexscreener.com/solana

# html_file_path = 'dexscreener_solana.html'
# with open(html_file_path, 'r', encoding='utf-8') as file:
#     html_content = file.read()
url = 'https://dexscreener.com/solana'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

data = []
headers = ['Token', 'Price', 'TXNS', 'Volume', 'Makers', '5M', '1H', '6H', '24H', 'Liquidity', 'MCAP']

rows = soup.select('a.ds-dex-table-row')

# Iterate over each row and extract the data
for row in rows:
    cells = row.find_all('div', class_='ds-table-data-cell')
    print(f"Number of cells in the row: {len(cells)}")
    if len(cells) == 11:  # Ensure the row has the expected number of columns
        # Extract the token name components and concatenate them
        base_token_symbol = cells[0].find('span', class_='ds-dex-table-row-base-token-symbol').get_text(strip=True)
        quote_token_symbol = cells[0].find('span', class_='ds-dex-table-row-quote-token-symbol').get_text(strip=True)
        base_token_name = cells[0].find('span', class_='ds-dex-table-row-base-token-name').get_text(strip=True)
        token_name = f"{base_token_symbol}/{quote_token_symbol} {base_token_name}"
        
        # Extract other data
        other_data = [cell.get_text(strip=True) for cell in cells[1:]]
        token_data = [token_name] + other_data
        data.append(token_data)

# Write the data to a CSV file
csv_file_path = 'solana_surge_tokens.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(headers)
    writer.writerows(data)
```

```python
# Scrape dexscreener.com/solana loop

data = []
headers = ['Token', 'Price', 'TXNS', 'Volume', 'Makers', '5M', '1H', '6H', '24H', 'Liquidity', 'MCAP']
base_url = 'https://dexscreener.com/solana'

page = 1
while True:
    url = f"{base_url}/page-{page}"  # Adjust URL based on actual parameter
    print(f"Fetching data from {url}")
    response = requests.get(url)
    html_content = response.content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    rows = soup.select('a.ds-dex-table-row')
    if not rows:  # Break the loop if no more rows are found
        break

    print(f"Number of rows found on page {page}: {len(rows)}")
    
    # Iterate over each row and extract the data
    for row in rows:
        cells = row.find_all('div', class_='ds-table-data-cell')
        print(f"Number of cells in the row: {len(cells)}")
        if len(cells) == 11:  # Ensure the row has the expected number of columns
            # Extract the token name components and concatenate them
            base_token_symbol = cells[0].find('span', class_='ds-dex-table-row-base-token-symbol').get_text(strip=True)
            quote_token_symbol = cells[0].find('span', class_='ds-dex-table-row-quote-token-symbol').get_text(strip=True)
            base_token_name = cells[0].find('span', class_='ds-dex-table-row-base-token-name').get_text(strip=True)
            token_name = f"{base_token_symbol}/{quote_token_symbol} {base_token_name}"
            
            # Extract other data
            other_data = [cell.get_text(strip=True) for cell in cells[1:]]
            token_data = [token_name] + other_data
            data.append(token_data)

    page += 1
    time.sleep(5)
    
    
# Write the data to a CSV file
csv_file_path = 'solana_surge_tokens_all.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(headers)
    writer.writerows(data)
```

```python
# Solana Surging Coin Monitoring Agent (SerperDevTool)

agent = Agent(
    role="Surging Coin Monitor",
    goal="""Monitor dexscreener.com/solana for the most surging coins currently on the Solana network. 
    This includes tracking coins by volume, number of transactions, and increase in market cap. 
    The goal is to alert you about the coins that are experiencing the highest surges. Ensure to include the current 
    market cap and volume for each coin in the final output. Exclude coins with miniscule/low liquidity or transaction volume.""",
    backstory=
        """As a dedicated crypto monitor, your job is to identify the most surging coins within the last few hours/days
        on the Solana network using data from dexscreener.com/solana. You analyze various metrics such as trading volume, 
        transaction counts, and market cap increases to find the coins that are gaining the most traction. Ensure to include the current 
        market cap and volume for each coin in the final output.""",
    verbose=True,
    allow_delegation=True,
    tools=[search_tool],
)
task = Task(
    description="Research a large number of the coins on dexscreener.com/solana and filter and find the most recent surging coins on the Solana network and provide their current market cap and volume and other informative details.",
    expected_output="Detailed report of the top 10 most surging coins found, including their names, trading volume, transaction counts, market cap increases, current market cap, and volume.",
    agent=agent,
    output_file='./logs/surging_coins_output.txt'
)

# Set up the crew to execute the task
crew = Crew(
    agents=[agent],
    tasks=[task]
)

# Run the crew and handle results
result = crew.kickoff()
```

```python
# Solana New Coin Monitoring Agent

agent = Agent(
    role="New Coin Monitor",
    goal="""Monitor dexscreener.com for new coins listed on the Solana network. 
    This includes checking the latest listings and providing detailed information 
    about each new coin. The goal is to alert you about any new listings.""",
    backstory=
        """As a dedicated crypto monitor, your job is to stay ahead of the curve by 
        identifying new coins listed on the Solana network using data from dexscreener.com. 
        You scan the latest listings to find new coins and ensure that no new coin listing goes unnoticed.""",
    verbose=True,
    allow_delegation=True,
    tools=[search_tool],
)

# Define the task for the agent
task = Task(
    description="Search dexscreener.com for new coins listed on the Solana network",
    expected_output="Detailed report of any new coins found, including their names and initial information.",
    agent=agent,
    output_file='./logs/new_coins_output.txt'
)

# Set up the crew to execute the task
crew = Crew(
    agents=[agent],
    tasks=[task],
    verbose=2
)

# Run the crew and handle results
result = crew.kickoff() 
print(result)
```
